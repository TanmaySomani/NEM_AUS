```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from math import sqrt

# Load the datasets
demand_actual = pd.read_csv('demand_actual.csv')
rooftop_solar = pd.read_csv('rooftop_solar.csv')
prices = pd.read_csv('prices.csv')

# Filter for NSW1 region in each dataset
nsw_demand = demand_actual[demand_actual['REGIONID'] == 'NSW1'].copy()
nsw_solar = rooftop_solar[rooftop_solar['REGIONID'] == 'NSW1'].copy()
nsw_prices = prices[prices['REGIONID'] == 'NSW1'].copy()

# Convert 'date_time' columns to datetime format
nsw_demand['date_time'] = pd.to_datetime(nsw_demand['date_time'])
nsw_solar['INTERVAL_DATETIME'] = pd.to_datetime(nsw_solar['INTERVAL_DATETIME'])
nsw_prices['date_time'] = pd.to_datetime(nsw_prices['date_time'])

# Merge datasets based on the date_time column
merged_data = pd.merge(nsw_demand[['date_time', 'OPERATIONAL_DEMAND']], 
                       nsw_solar[['INTERVAL_DATETIME', 'POWER']], 
                       left_on='date_time', right_on='INTERVAL_DATETIME', how='inner')
merged_data = pd.merge(merged_data, nsw_prices[['date_time', 'RRP']], on='date_time', how='inner')


# Drop any unnecessary columns and NaN values
merged_data.drop(columns=['INTERVAL_DATETIME'], inplace=True)
merged_data.dropna(inplace=True)

# Visualize the merged data (Optional)
merged_data
```

```python


# Normalize the data (demand, solar, and price)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(merged_data[['OPERATIONAL_DEMAND', 'POWER', 'RRP']])

# Create a supervised learning problem (with look-back windows)
def create_dataset(dataset, look_back=60):
    X, y = [], []
    for i in range(len(dataset) - look_back):
        X.append(dataset[i:(i + look_back), :])
        y.append(dataset[i + look_back, 0])  # We only predict the demand
    return np.array(X), np.array(y)

look_back = 60  # 60 time steps
X, y = create_dataset(scaled_data, look_back)

# Split into training and testing sets (80% train, 20% test)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the LSTM model (with additional features: solar and price)
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=25))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=30)

# Make predictions
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(np.concatenate([predictions, np.zeros((predictions.shape[0], 2))], axis=1))[:,0]

# Inverse transform the y_test to actual values
y_test_actual = scaler.inverse_transform(np.concatenate([y_test.reshape(-1, 1), np.zeros((y_test.shape[0], 2))], axis=1))[:,0]

# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test_actual, predictions))
print(f'RMSE: {rmse}')

# Plot predictions vs actual values
plt.figure(figsize=(10,6))
plt.plot(y_test_actual, label='Actual Demand')
plt.plot(predictions, label='Predicted Demand', color='red')
plt.title('NSW1 Demand Forecast vs Actual (LSTM with Solar and Price)')
plt.xlabel('Time')
plt.ylabel('Operational Demand')
plt.legend()
plt.show()
```

```python
# Calculate percentage change in predicted demand
predicted_pct_change = np.diff(predictions.reshape(-1)) / predictions[:-1].reshape(-1)

# Define a threshold for spike detection (e.g., 10% increase)
spike_threshold = 0.10
spikes = predicted_pct_change > spike_threshold

# Get the spike indices and times
spike_indices = np.where(spikes)[0]
print("Predicted Demand Spikes at Test Set Indices:")
print(spike_indices)

# Plot the spikes on the demand forecast
plt.figure(figsize=(10,6))
plt.plot(y_test_actual, label='Actual Demand')
plt.plot(predictions, label='Predicted Demand', color='red')

# Highlight the spike points
plt.scatter(spike_indices, predictions[spike_indices], color='green', label='Spikes', marker='x')
plt.title('NSW1 Demand Forecast with Spikes (LSTM)')
plt.xlabel('Time')
plt.ylabel('Operational Demand')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Make a copy of predictions to adjust for redistribution
adjusted_predictions = predictions.copy()

# Define the redistribution percentage (e.g., 10%)
redistribution_percentage = 0.5

# Define the window over which to redistribute (e.g., previous 3 hours)
redistribution_window = 12

# Adjust predictions to redistribute demand during spikes
for spike_idx in spike_indices:
    if spike_idx >= redistribution_window:
        # Calculate the amount to redistribute
        demand_to_redistribute = adjusted_predictions[spike_idx] * redistribution_percentage
        # Subtract the demand from the spike
        adjusted_predictions[spike_idx] -= demand_to_redistribute
        # Evenly distribute it over the previous `redistribution_window` hours
        for i in range(1, redistribution_window + 1):
            adjusted_predictions[spike_idx - i] += demand_to_redistribute / redistribution_window

# Plot original predictions and adjusted predictions for comparison
plt.figure(figsize=(10,6))
plt.plot(y_test_actual, label='Actual Demand')
plt.plot(predictions, label='Original Predicted Demand', color='red')
plt.plot(adjusted_predictions, label='Adjusted Predicted Demand', color='green')

# Highlight the original spikes
plt.scatter(spike_indices, predictions[spike_indices], color='orange', label='Original Spikes', marker='x')

plt.title('NSW1 Demand Forecast with Adjusted Predictions')
plt.xlabel('Time')
plt.ylabel('Operational Demand')
plt.legend()
plt.show()
```

```python
# Define the downsampling factor (e.g., average every 10 points)
downsample_factor = 100

# Downsample the actual demand, predictions, and adjusted predictions
y_test_actual_downsampled = y_test_actual[:len(y_test_actual) - len(y_test_actual) % downsample_factor].reshape(-1, downsample_factor).mean(axis=1)
predictions_downsampled = predictions[:len(predictions) - len(predictions) % downsample_factor].reshape(-1, downsample_factor).mean(axis=1)
adjusted_predictions_downsampled = adjusted_predictions[:len(adjusted_predictions) - len(adjusted_predictions) % downsample_factor].reshape(-1, downsample_factor).mean(axis=1)

# Downsample the spike indices (adjust to match downsampling)
spike_indices_downsampled = (spike_indices / downsample_factor).astype(int)

# Plot the downsampled predictions for comparison
plt.figure(figsize=(10,6))
plt.plot(y_test_actual_downsampled, label='Actual Demand (Downsampled)')
plt.plot(predictions_downsampled, label='Original Predicted Demand (Downsampled)', color='red')
plt.plot(adjusted_predictions_downsampled, label='Adjusted Predicted Demand (Downsampled)', color='green')

# Highlight the original spikes
plt.scatter(spike_indices_downsampled, predictions_downsampled[spike_indices_downsampled], color='orange', label='Original Spikes', marker='x')

plt.title('NSW1 Demand Forecast (Downsampled) with Adjusted Predictions')
plt.xlabel('Time')
plt.ylabel('Operational Demand')
plt.legend()
plt.show()
```
